```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
import nltk
```

```python
story = []
for filename in os.listdir(r'C:\Users\lshre\data'):
    if filename == '.ipynb_checkpoints':
      pass
    f = open(os.path.join(r'C:\Users\lshre\data',filename))
    corpus = f.read()
    raw_sent = sent_tokenize(corpus)
    for sent in raw_sent:
        story.append(simple_preprocess(sent))
```

```python
story
```

```python
story[0]
```

```python
model = gensim.models.Word2Vec(
    window=10,
    min_count=2
)
```

```python
model.build_vocab(story)
```

```python
model.train(story, total_examples=model.corpus_count, epochs=model.epochs)
```

```python

from sklearn.decomposition import PCA #Principal Component Analysis
```

```python

pca = PCA(n_components=3)
```

```python

X = pca.fit_transform(model.wv.get_normed_vectors())
```

```python
X
```

```python
X.shape
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Create scatter plot with smaller points and some transparency
scatter = ax.scatter(X[:, 0], 
                    X[:, 1], 
                    X[:, 2],
                    c=X[:, 2],  # Color by z-coordinate
                    cmap='viridis',
                    alpha=0.6,
                    s=5)  # Smaller point size

# Labels and title
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')
ax.set_title('Interactive 3D Visualization of Text Vectors\n(Click and drag to rotate)')

plt.colorbar(scatter)

# Enable interactive rotation
def rotate(event):
    ax.view_init(elev=ax.elev + (event.ydata - ax._prev_ydata) * 0.3,
                 azim=ax.azim + (event.xdata - ax._prev_xdata) * 0.3)
    ax._prev_xdata = event.xdata
    ax._prev_ydata = event.ydata
    plt.draw()

def on_mouse_press(event):
    ax._prev_xdata = event.xdata
    ax._prev_ydata = event.ydata

fig.canvas.mpl_connect('motion_notify_event', rotate)
fig.canvas.mpl_connect('button_press_event', on_mouse_press)

plt.show()
```
